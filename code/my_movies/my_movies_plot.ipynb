```python
import pandas as pd
from matplotlib import pyplot as plt

years_df = pd.read_csv('../../data/my/years.csv')
writers_df = pd.read_csv('../../data/my/writers.csv')
runtimeInterval_df = pd.read_csv('../../data/my/runtimeInterval.csv')
numVotesInterval_df = pd.read_csv('../../data/my/numVotesInterval.csv')
nconst_df = pd.read_csv('../../data/my/nconst.csv')
genres_df = pd.read_csv('../../data/my/genres.csv')
directors_df = pd.read_csv('../../data/my/directors.csv')
decades_df = pd.read_csv('../../data/my/decades.csv')

plt.style.use('ggplot')
```

```python
years_x = years_df['startYear']
average_rating_year_y = years_df['averageRating_mean']
plt.plot(years_x, average_rating_year_y)
plt.title('Average Movie Rating by Year')
plt.xlabel('Years')
plt.ylabel('Average Movie Rating')
plt.show()
```

```python
year_count_y = years_df['count']
plt.plot(years_x, year_count_y)
plt.title('Released Films by Year')
plt.xlabel('Years')
plt.ylabel('Released Films')
plt.show()
```

```python
decades_x = decades_df['decade']
average_rating_decade_y = decades_df['averageRating_mean']
plt.plot(decades_x, average_rating_decade_y)
plt.title('Average Movie Rating by Decade')
plt.xlabel('Decades')
plt.ylabel('Average Rating')
plt.show()
```

```python
runtimeInterval_x = runtimeInterval_df['runtimeInterval']
average_rating_runtimeInterval_y = runtimeInterval_df['averageRating_mean']
plt.plot(runtimeInterval_x, average_rating_runtimeInterval_y)
plt.title('Average Movie Rating by RunTime Interval')
plt.xlabel('RunTime Interval')
plt.ylabel('Average Movie Rating')
plt.show()
```
