```python
import pandas as pd
from matplotlib import pyplot as plt

years_df = pd.read_csv('../../data/my/years.csv')
writers_df = pd.read_csv('../../data/my/writers.csv')
runtimeInterval_df = pd.read_csv('../../data/my/runtimeInterval.csv')
numVotesInterval_df = pd.read_csv('../../data/my/numVotesInterval.csv')
nconst_df = pd.read_csv('../../data/my/nconst.csv')
genres_df = pd.read_csv('../../data/my/genres.csv')
directors_df = pd.read_csv('../../data/my/directors.csv')
decades_df = pd.read_csv('../../data/my/decades.csv')

plt.style.use('ggplot')
```

```python
decades_x = decades_df['decade']
average_rating_decade_y = decades_df['averageRating_mean']
plt.bar(decades_x, average_rating_decade_y, width=2)
plt.plot(decades_x, average_rating_decade_y, color='blue',linestyle='--' )
plt.title('Personal Score by Decade')
plt.xlabel('Decades')
plt.ylabel('Personal Score')
plt.show()
```

```python
runtimeInterval_x = runtimeInterval_df['runtimeInterval']
average_rating_runtimeInterval_y = runtimeInterval_df['averageRating_mean']
plt.bar(runtimeInterval_x, average_rating_runtimeInterval_y)
plt.title('Personal Score by RunTime Interval')
plt.xlabel('RunTime Interval')
plt.ylabel('Personal Score')
plt.show()
```

```python
genres_df = genres_df.sort_values(by='count')
genres_x = genres_df['genres']
count_y = genres_df['count']
plt.barh(genres_x, count_y)
plt.title('Films watched by genre')
plt.xlabel('Genres')
plt.show()
```

```python
genres_df = genres_df.sort_values(by='personal_score_mean')
genres_x = genres_df['genres']
personal_score_mean_y = genres_df['personal_score_mean']
plt.barh(genres_x, personal_score_mean_y)
plt.title('Genres by personal score')
plt.xlabel('Personal Score')
plt.show()
```

```python
directors_count_df = directors_df.sort_values(by='count', ascending=False)
directors_x = directors_count_df['directors'].head(10)
count_y = directors_count_df['count'].head(10)
plt.barh(directors_x, count_y)
plt.title('TOP-10 most watched directors')
plt.gca().invert_yaxis()
plt.show()
```

```python
directors_score_df = directors_df.loc[directors_df['count']>5].sort_values(by='personal_score_mean')
directors_x = directors_score_df['directors']
personal_score_mean_y = directors_score_df['personal_score_mean']
plt.barh(directors_x, personal_score_mean_y)
plt.title('TOP Directors by personal score')
plt.xlabel('Personal Score')
plt.show()
```

```python
writers_count_df = writers_df.sort_values(by='count', ascending=False)
writers_x = writers_count_df['writers'].head(10)
count_y = writers_count_df['count'].head(10)
plt.barh(writers_x, count_y)
plt.title('TOP-10 most watched writers')
plt.gca().invert_yaxis()
plt.show()
```

```python
writers_score_df = writers_df.loc[writers_df['count']>5].sort_values(by='personal_score_mean')
writers_x = writers_score_df['writers']
personal_score_mean_y = writers_score_df['personal_score_mean']
plt.barh(writers_x, personal_score_mean_y)
plt.title('TOP writers by personal score')
plt.xlabel('Personal Score')
plt.show()
```

```python
nconst_count_df = nconst_df.sort_values(by='count', ascending=False)
nconst_x = nconst_count_df['nconst'].head(10)
count_y = nconst_count_df['count'].head(10)
plt.barh(nconst_x, count_y)
plt.title('TOP-10 most watched actors')
plt.gca().invert_yaxis()
plt.show()
```

```python
nconst_score_df = nconst_df.loc[nconst_df['count']>5].sort_values(by='personal_score_mean', ascending=False)
nconst_x = nconst_score_df['nconst'].head(10)
personal_score_mean_y = nconst_score_df['personal_score_mean'].head(10)
plt.barh(nconst_x, personal_score_mean_y)
plt.title('TOP actors by personal score')
plt.xlabel('Personal Score')
plt.gca().invert_yaxis()
plt.show()
```
